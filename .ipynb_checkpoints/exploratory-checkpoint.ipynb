### Imbalanced Dataset, predicting Shootings being invovled with crimes in Boston

```python
df.Shooting.value_counts()
```

There are 58208 counts of crimes where a shooting wasn't invovled. <br>
There are 636 counts of crimes where a shooting was involved.

```python
sns.countplot(x= 'Shooting', data= df)
```

```python
df.info()
```

```python
df = df.drop(['id', 'Description', 'Area', 'Date'], 1)
```

```python
df = df.drop('Year', 1) 
```

```python
group_train = pd.get_dummies(df['Group'], prefix='grp', prefix_sep='_')
district_train = pd.get_dummies(df['District'], prefix='district', prefix_sep='_')
month_train = pd.get_dummies(df['Month'], prefix='month', prefix_sep='_')
day_train = pd.get_dummies(df['Day'], prefix='day', prefix_sep='_')
hour_train = pd.get_dummies(df['Hour'], prefix='hour', prefix_sep='_')
```

```python
def shoot_bin(df):
    if df.Shooting == 'Y':
        return 1
    elif df.Shooting == 'N':
        return 0 
    else:
        return df.Shooting
```

```python
df['Shooting'] = df.apply(shoot_bin, 1)
```

```python
df = df.drop(['Group', 'District', 'Month', 'Day', 'Hour'], 1)
```

```python
df_final = pd.concat([df,
                          group_train,
                          district_train,
                          month_train,
                          day_train,
                          hour_train
                         ], axis=1)
```

```python
df_final.shape
```

### The Dangers of Imbalanced Datasets

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
X = df_final.drop('Shooting', 1)
y = df_final['Shooting']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=21, stratify=y)
```

```python
model1 = LogisticRegression(solver = 'lbfgs')
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
print('Accuracy of the model is : ' + str(model1.score(X_test, y_test)))
print('-----')
count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
print('-----')
print('The Confusion Matrix :')
print(confusion_matrix(y_test,y_pred ))
print('-----')
print('Classification Report :')
print(classification_report(y_test, y_pred))
```

The accuracy looks good at **99%**, but that is because the majority of the predictions were 0 to begin with. Even if the model predicted all 0's the model would still be above 99%+ accuracy. 

We can see that the recall rate of predictions for shootings being 1's are 0.2! **Only 20% of cases where there was actually a shooting was correctly identified.**

We can see the dangers of imbalanced datasets from this model prediciton. We would like to know when a Shooting will be involved in a crime, so let's try to "balance out" the dataset.


Sub-sampling = our subsample will be a dataframe with a 50/50 ratio of crimes with shootings and crimes without shootings.

We subsample because otherwise, the classification models would overfit and assume that in most cases there are no shootings. What we actually want is for our model to be certain of when a shooting does occur. 

We also wouldn't be able to understand how each of these features infulence the result, and we wouldn't be able to see the true correlations between the class and features (in shooting cases).

### Random Under Sampling

We implement a random under sampling technique to arrive at a balanced dataset. However, this runs the risk that our classificaiton models will not perform as accurately as we wouldd like, since there is a great deal of information loss (bringing 636 non shooting cases from 58k non shooting cases)

```python
from sklearn.utils import resample

#separate teh majority and minority classes
df_majority = df_final[df_final.Shooting == 0]
df_minority = df_final[df_final.Shooting == 1]

#downsample the majority class
df_majority_downsampled = resample(df_majority,
                                  replace = False,  #sample w/0 replacement
                                  n_samples = 636,  #match minority class #
                                  random_state = 42)#reproducible results

#combine minority class with the downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

#display the new class counts
df_downsampled.Shooting.value_counts()
```

```python
sns.countplot(x='Shooting', data= df_downsampled)
```

```python
# Make sure we use the subsample in our correlation

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(24,20))

# Entire DataFrame
corr = df_final.corr()
sns.heatmap(corr, cmap='coolwarm_r', annot_kws={'size':20}, ax=ax1)
ax1.set_title("Imbalanced Correlation Matrix \n (don't use for reference)", fontsize=14)


sub_sample_corr = df_downsampled.corr()
sns.heatmap(sub_sample_corr, cmap='coolwarm_r', annot_kws={'size':20}, ax=ax2)
ax2.set_title('SubSample Correlation Matrix \n (use for reference)', fontsize=14)
plt.show()
```

We can see from the correlation matrices above that in both the actual data and the subsampled data that the time doesn't have much correlation within the features. All of the categorical variables for month, day, and hour seem to be weak. 

The new DataFrame has fewer observations than the original, and the ratio of the two classes is now 1:1.

### Dimensionality reduction and clustering:

t-SNE: intro-video https://www.youtube.com/watch?v=NEaUSP4YerM

- t-SNE algorithm can accurately cluster cases where there were shooting and non shooting cases in our dataset.
- although the random downsampled subset is pretty small (around 600 cases each), the t-SNE algorithm detects clusters in almost every case.
- This give an indication that furhter predictive models will perform well in separating shooting cases from non shooting cases.

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA, TruncatedSVD
import time

#using random undersample data 
X = df_downsampled.drop('Shooting', 1)
y = df_downsampled['Shooting']


# T-SNE Implementation
t0 = time.time()
X_reduced_tsne = TSNE(n_components=2, random_state=42).fit_transform(X.values)
t1 = time.time()
print("T-SNE took {:.2} s".format(t1 - t0))

# PCA Implementation
t0 = time.time()
X_reduced_pca = PCA(n_components=2, random_state=42).fit_transform(X.values)
t1 = time.time()
print("PCA took {:.2} s".format(t1 - t0))

# TruncatedSVD
t0 = time.time()
X_reduced_svd = TruncatedSVD(n_components=2, algorithm='randomized', random_state=42).fit_transform(X.values)
t1 = time.time()
print("Truncated SVD took {:.2} s".format(t1 - t0))
```

```python
import matplotlib.patches as mpatches

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24,6))

f.suptitle('Clusters using Dimensionality Reduction', fontsize=14)


blue_patch = mpatches.Patch(color='#0A0AFF', label='No Shooting')
red_patch = mpatches.Patch(color='#AF0000', label='Shooting')


# t-SNE scatter plot
ax1.scatter(X_reduced_tsne[:,0], X_reduced_tsne[:,1], c=(y == 0), cmap='coolwarm', label='No Fraud', linewidths=2)
ax1.scatter(X_reduced_tsne[:,0], X_reduced_tsne[:,1], c=(y == 1), cmap='coolwarm', label='Fraud', linewidths=2)
ax1.set_title('t-SNE', fontsize=14)

ax1.grid(True)

ax1.legend(handles=[blue_patch, red_patch])


# PCA scatter plot
ax2.scatter(X_reduced_pca[:,0], X_reduced_pca[:,1], c=(y == 0), cmap='coolwarm', label='No Fraud', linewidths=2)
ax2.scatter(X_reduced_pca[:,0], X_reduced_pca[:,1], c=(y == 1), cmap='coolwarm', label='Fraud', linewidths=2)
ax2.set_title('PCA', fontsize=14)

ax2.grid(True)

ax2.legend(handles=[blue_patch, red_patch])

# TruncatedSVD scatter plot
ax3.scatter(X_reduced_svd[:,0], X_reduced_svd[:,1], c=(y == 0), cmap='coolwarm', label='No Fraud', linewidths=2)
ax3.scatter(X_reduced_svd[:,0], X_reduced_svd[:,1], c=(y == 1), cmap='coolwarm', label='Fraud', linewidths=2)
ax3.set_title('Truncated SVD', fontsize=14)

ax3.grid(True)

ax3.legend(handles=[blue_patch, red_patch])

plt.show()
```

While the accuracy of the overall model went down to **92.9%**, the recall rate of the true shooting cases has significantly increased to **95%**. This recall metric carries much more weight, knowing which crimes involve a shooting is what we are actually interested in.

Another performance metric that we can look at is the **Area Under ROC Curve**.

Intuitively, the AUROC represents the likelihood of your model distinguishing observations from the two classes. In other words, if you reandomly select one observation from each class, what's the probability that your model will be able to rank them correctly.

```python
from sklearn.metrics import roc_auc_score
# Predict class probabilities
prob_y_2 = model2.predict_proba(X)
 
# Keep only the positive class
prob_y_2 = [p[1] for p in prob_y_2]
 
prob_y_2[:5] 
```

The logistic regression model trained on the downsampled dataset has a 98% probability of distinguishing a random observation from the dataset. 

```python
print( roc_auc_score(y, prob_y_2) )
```

On the other hand, the original model trained on the imbalanced dataset has the following AUROC score of 97.7%. The model trained on the random downsampled balanced dataset performs slightly better.

```python
prob_y_0 = model1.predict_proba(X)
prob_y_0 = [p[1] for p in prob_y_0]
 
print( roc_auc_score(y, prob_y_0) )
```

## Let's compare some different classification models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import collections
from collections import Counter

from sklearn.pipeline import make_pipeline
```

```python
X = df_downsampled.drop('Shooting', 1)
y = df_downsampled['Shooting']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=21, stratify=y)
```

```python
# Turn the values into an array for feeding the classification algorithms.
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
#simple classifiers

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier()
}
```

```python
from sklearn.model_selection import cross_val_score

for key, classifier in classifiers.items():
    classifier.fit(X_train, y_train)
    training_score = cross_val_score(classifier, X_train, y_train, cv=5)
    print("Classifiers: ", classifier.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```

```python
# Use GridSearchCV to find the best parameters.
from sklearn.model_selection import GridSearchCV


# Logistic Regression 
log_reg_params = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}



grid_log_reg = GridSearchCV(LogisticRegression(), log_reg_params)
grid_log_reg.fit(X_train, y_train)
# We automatically get the logistic regression with the best parameters.
log_reg = grid_log_reg.best_estimator_

knears_params = {"n_neighbors": list(range(2,5,1)), 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}

grid_knears = GridSearchCV(KNeighborsClassifier(), knears_params)
grid_knears.fit(X_train, y_train)
# KNears best estimator
knears_neighbors = grid_knears.best_estimator_

# Support Vector Classifier
svc_params = {'C': [0.5, 0.7, 0.9, 1], 'kernel': ['rbf', 'poly', 'sigmoid', 'linear']}
grid_svc = GridSearchCV(SVC(), svc_params)
grid_svc.fit(X_train, y_train)

# SVC best estimator
svc = grid_svc.best_estimator_

# DecisionTree Classifier
tree_params = {"criterion": ["gini", "entropy"], "max_depth": list(range(2,4,1)), 
              "min_samples_leaf": list(range(5,7,1))}
grid_tree = GridSearchCV(DecisionTreeClassifier(), tree_params)
grid_tree.fit(X_train, y_train)

# tree best estimator
tree_clf = grid_tree.best_estimator_
```

```python
# Overfitting Case

log_reg_score = cross_val_score(log_reg, X_train, y_train, cv=5)
print('Logistic Regression Cross Validation Score: ', round(log_reg_score.mean() * 100, 2).astype(str) + '%')

knears_score = cross_val_score(knears_neighbors, X_train, y_train, cv=5)
print('Knears Neighbors Cross Validation Score', round(knears_score.mean() * 100, 2).astype(str) + '%')

svc_score = cross_val_score(svc, X_train, y_train, cv=5)
print('Support Vector Classifier Cross Validation Score', round(svc_score.mean() * 100, 2).astype(str) + '%')

tree_score = cross_val_score(tree_clf, X_train, y_train, cv=5)
print('DecisionTree Classifier Cross Validation Score', round(tree_score.mean() * 100, 2).astype(str) + '%')
```
